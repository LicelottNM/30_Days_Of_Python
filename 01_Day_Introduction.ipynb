```python
# Use basic arithmetic operators

print(2 + 2) # Addition
print(4 - 2) # Subtraction
print(2 * 2) # Multiplication
print(2 / 2) # Division
print(2 ** 2) # Exponentiation -> 2 to the power of 2
print(2 % 5) # Modulo -> Remainder of 2 divided by 5
print(2 // 2) # Floor division -> Quotient of 2 divided by 2, rounded down to the nearest whole number
```

```python
# Checking data types
print(type(2)) # int -> Integer
print(type(2.33)) # float -> Floating-point number
print(type(1 + 3j)) # complex -> Complex number
print(type('hi mom')) #string -> String
print(type([1, 2, 3])) #list -> List
print(type({'name': 'Licelott'})) # Dictionary -> Dictionary
print(type({9.8, 3.14, 45.6})) #tuple -> Tuple
```
